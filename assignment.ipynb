# Model 2 CNN


### I've create a simple Convolutional Neural Network (CNN) architecture which is different from the simple dense network normally used in model 1

##What the model does:

####The CNN model takes as input a set of 2D images (28x28 pixels). Each image represents an article of clothing, such as a shirt, a shoe, a bag, etc. The goal of the model is to correctly identify the type of clothing present in each image. It does this by learning from the pixel data in the images and the corresponding labels provided during the training phase

```python
# Import necessary modules
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

##How the input/output patterns were chosen:


###The input to this model is an image, represented as a 2D array of pixel intensity values. The output is a probability distribution across the 10 different classes (types of clothing), indicating the model's confidence that the image belongs to each particular class. The model will then predict the class with the highest probability. The input/output pattern is a typical setup for image classification tasks

###Model architecture/structure

```python
# Reshape the images to include a channel dimension (needed for Conv2D layers)
train_images = train_images.reshape(-1, 28, 28, 1)
test_images = test_images.reshape(-1, 28, 28, 1)

# Define the model architecture
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

###In this architecture, the model starts with a series of convolutional and pooling layers, followed by a flattening step, and ends with two dense layers. The convolutional layers are responsible for detecting patterns in the images, while the dense layers interpret these patterns and make a final prediction.

```python
# Add Dense layers on top
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
history = model.fit(train_images, train_labels, epochs=10,
                    validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

### Note that shifting from simple dense layer to CNN model the accuracy increased from .88 to.90
